**1.2. Print the first 100 base pairs, then 1000bp. Store both chunks of DNA into a variable.**

**1.3. Let's count the number of every nucleotide in each chunk.**

```python
short_count={"A":0, "C":0, "T":0, "G":0}
for base in seq_100:
    short_count[base]+=1

print(short_count)
```

```python
long_count={"A":0, "C":0, "T":0, "G":0}
for base in seq_1000:
    long_count[base]+=1
print(long_count)
```

**1.4. Estimate the nucleotide frequencies in both chunks.**

```python
short_count_freq=short_count
for base in short_count_freq:
    short_count_freq[base]=short_count_freq[base]/len(seq_100)
print(short_count_freq)
```

```python
long_count_freq=long_count
for base in long_count_freq:
    long_count_freq[base]=long_count_freq[base]/len(seq_1000)
print(long_count_freq)
```

**2.1. Let's have a look at the variation in GC content along the gene body**
    *Let's have a look at the variation in GC content along the gene body Choose an appropriate window size and plot the results per-window. (Use pyplot, dictionaries to store base counts, range funtion for non-overlapping windows. Note that matplotlib has to be installed first for example using conda)*


```python

import matplotlib.pyplot as plt


def GCplotter(gene, window):
    coordinate = []
    GCcontent = []

    for position in range(window, len(gene), window):

        my_chunk=gene[(position-window):position]
        CHUNKcounts = {'A': 0, 'T': 0, 'G': 0, 'C':0}
        for base in my_chunk:
            CHUNKcounts[base] += 1
        coordinate.append(position)
        GCcontent.append(((CHUNKcounts.get("G")+CHUNKcounts.get("C"))/len(my_chunk)))

    plt.plot(coordinate, GCcontent)
    plt.ylabel('GC content')
    plt.xlabel('Position (bp)')
    plt.show()


window = 500

print(GCplotter(ACE2, window))
```

```python
file="PPT1_plus_promoter_1kb_uppercase.fa"
filename="/home/jj/Desktop/Bioinformatics/2nd_year/2term/ASAB/Seminars/S3_Estimating_GC_content/data/"+file
read=SeqIO.read(filename, "fasta")

gene = read.seq
window = 500

print(GCplotter(gene, window))
```

```python
import math
coordinate = []
ratio = []


for position in range(window, len(gene), window):

    my_chunk=gene[(position-window):position]
    CHUNKcounts = {'A': 0, 'T': 0, 'G': 0, 'C':0}
    for base in my_chunk:
        CHUNKcounts[base] += 1
    coordinate.append(position)
    GCcontent=(CHUNKcounts.get("G")+CHUNKcounts.get("C"))/len(my_chunk)
    ATcontent=(CHUNKcounts.get("A")+CHUNKcounts.get("T"))/len(my_chunk)
    ratio.append(math.log(GCcontent/ATcontent))

plt.plot(coordinate, ratio)
plt.ylabel('content')
plt.xlabel('Position (bp)')
plt.show()
```
